```python
# 판다스
import pandas as pd
# requests
import requests as req
# bs
from bs4 import BeautifulSoup as bs

# header
header = {'user-agent':'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/109.0.0.0 Safari/537.36'}
```

```python
# 장르
genre = 'GN0100' # 발라드
# 페이지
page = '1' # 1페이지 1위 ~ 50위

# 장르, 페이지 이용한 멜론차트 url 요청
url = f'https://www.melon.com/genre/song_list.htm?gnrCode={genre}#params%5BgnrCode%5D=GN0100&params%5BdtlGnrCode%5D=&params%5BorderBy%5D=NEW&params%5BsteadyYn%5D=N&po=pageObj&startIndex={page}'
res = req.get(url, headers = header)
# 파이썬 객체변환
html = bs(res.text, 'lxml')
```

```python
# 곡명 크롤링
song = html.select('.ellipsis.rank01>span')
song_lst = [i.text.strip() for i in song]
song_lst
```

```python
# 가수명 크롤링
singer = html.select('.ellipsis.rank02>span')
singer_lst = [i.text.strip() for i in singer]
singer_lst
```

```python
# 곡명, 가수명 딕셔너리 정의
GN0100_dict = {'곡명':song_lst , '가수명':singer_lst}
```

```python
# 데이터 프레임 구성
GN0100_df = pd.DataFrame(GN0100_dict)
```
